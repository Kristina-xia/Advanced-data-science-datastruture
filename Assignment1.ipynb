```python
from random import choice
from collections import Counter

days= list(range(1,366))

count= 0
for i in range(10000):
    persons = []
    for j in range(1,614):
        persons.append(choice(days))
        dayCount= Counter(persons)
    if dayCount[103]>=2:
        count += 1
print('Expected Result:%d'%(1/2*10000),'Simulated Result:%d'%count,sep='\n')
```

```python
from random import choice
from collections import Counter

coin=['head','back']

count=0
for i in range(10000):
    coins = []
    for j in range(1,4):
        coins.append(choice(coin))
    coinCount= Counter(coins)
    if coinCount['head']==2 and coinCount['back']==1:
        count += 1
print('Expected Result:%d'%(3/8*10000),'Simulated Result:%d'%count,sep='\n')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

n=3
p=0.5
k=np.arange(0,4)
binomial = stats.binom.pmf(k,n,p)

plt.plot(k,binomial,'o-')
plt.title('Binomial:n=%i, p=%.2f'%(n,p),fontsize=20)
plt.xlabel('Number of Successes')
plt.ylabel('Probability of Successes',fontsize=20)
plt.show()
```
